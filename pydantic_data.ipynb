```python
from pydantic import BaseModel,Field
from typing import List,Annotated
class User(BaseModel):
    name: str 
    age: int 
```

```python
user=User(name="John Doe", age=30)
user
```

```python
## Field Validation

class Product(BaseModel):
    name: str =Field(max_length=10)
    dis:str=Field(min_length=10)
    price: float=Field(gt=0,le=100)
```

```python
product=Product(name="Laptop", dis="This is a laptop", price=10)
product2=Product(name="Laptop", dis="This is a laptop", price=99)
```

```python
class Order(BaseModel):
    id: str=Field(min_length=5,max_length=10)
    product: Product
```

```python
order=Order(id="12345", product=product)
order
```

```python
order.product.price
```

```python
class Carts(BaseModel):
    id: str=Field(min_length=5,max_length=10)
    items: List[Product]
    comment:str=Field(
        default="No comment",
        max_length=100
    )
```

```python
cart=Carts(id="54321", items=[product,product2], comment="This is my cart")
cart
```

```python
for prod in cart.items:
    print(prod.name, prod.price)
```

```python
# Annotated - Helps to create common validator
```

```python
common=Annotated[List[Product], Field(default=[])]
```

```python
class Carts(BaseModel):
    id: str=Field(min_length=5,max_length=10)
    items: common
    comment:str=Field(
        default="No comment",
        max_length=100
    )
```

```python
class Order(BaseModel):
    id: str=Field(min_length=5,max_length=10)
    product: common
```

```python
ord=Order(id="12345", product=[product,product2])
ord
```
