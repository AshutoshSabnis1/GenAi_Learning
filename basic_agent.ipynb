```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
# Normal_way 
@tool
def add_numbers(a: int, b: int) -> int:
    """Add two numbers together.
    Args:
        a: The first number.
        b: The second number.  """
    return a + b

@tool
def multiply_numbers(a: int, b: int) -> int:
    """Multiply two numbers together.
    Args:
        a: The first number.
        b: The second number.  """
    return a * b

```

```python
add_numbers.invoke({"a": 1, "b": 2})
```

```python
# Agents_Way
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
agent=create_agent(
    model=llm,
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a helpful assistant that can perform basic arithmetic operations like addition and multiplication. Use the provided tools to answer questions about adding or multiplying numbers."
)
```

```python
response=agent.invoke({"messages": [{"role": "user", "content": "What is 3+5*10 ?"}]})
```
