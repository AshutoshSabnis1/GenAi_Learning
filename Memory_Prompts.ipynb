```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts=[{"role":"user","content":"My name is Ashutosh Sabnis . I Love GenAi"},
         {"role":"ai","content":"Hello Ashutosh ! How can I help you today ?"},
         {"role":"user","content":"What is my name ?"}]
res=llm.invoke(prompts)
print(res.content)
```

```python
# Manage History
history=[]
while True:
    query=input("User: ")
    if query.lower() in ["exit","quit"]:
        print("Goodbye!")
        break
    history.append({"role":"user","content":query})
    print("User: ",query)
    res=llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print("AI: ",res.content)
```
