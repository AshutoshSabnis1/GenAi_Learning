```python
# Prompt Templates (Dynamic Prompts)
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
out=StrOutputParser()  # By using this we don't write .content every time
prompts= ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a helpful assistant that translates English to {lang}."},
    {"role": "user", "content": "Translate: {query}"},])
chains=prompts|model|out 
response=chains.invoke({"lang": "Hindi", "query": "My name is Ashutosh Sabnis"})
print(response)
```
