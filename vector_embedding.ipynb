```python
from dotenv import load_dotenv
load_dotenv()
```

```python
documents=["Machine learning is a field of artificial intelligence that allows computers to learn from data.",
    "It helps systems make predictions or decisions without being explicitly programmed.",
    "Machine learning is widely used in applications like recommendation systems and image recognition.",
    "Common types of machine learning include supervised learning, unsupervised learning, and reinforcement learning.",
    "With more data and better algorithms, machine learning models can improve their accuracy over time."]
```

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings=OllamaEmbeddings(model="mxbai-embed-large")
```

```python
vectors=embeddings.embed_documents(documents)
len(vectors[0])
```

```python
# Store Vector Embeddings in ChromaDB
```

```python
from langchain_chroma import Chroma
```

```python
vector_store=Chroma.from_texts(
    texts=documents,
    embedding=embeddings,
    persist_directory="./chroma_db"
)
```

```python
query="Machine learning"
result=vector_store.similarity_search(query=query,k=2)
result
```
