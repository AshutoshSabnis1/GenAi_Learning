```python
## Document Text Loader
from langchain_community.document_loaders import TextLoader,PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader=TextLoader(r"c:\Users\ashut\OneDrive\Desktop\Machine Learning.txt")
texts=loader.load()
```

```python
pdfloader=PyPDFLoader(r"C:\Users\ashut\Downloads\Gen AI Workflow Specialist-Engineer.pdf")
pdfs=pdfloader.load()
len(pdfs)
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
chunks=splitter.split_documents(texts)
len(chunks)
```

```python
pdf_chunks=splitter.split_documents(pdfs)
len(pdf_chunks)
```
